```python
import cv2
import os
```

Read

```python
child_img = cv2.imread('child.png')
child_img.shape
```

```python
import matplotlib.pyplot as plt
```

Showing the Image

```python
plt.imshow(child_img)
```

```python
child_recolor = cv2.cvtColor(child_img, cv2.COLOR_BGR2RGB)
plt.imshow(child_recolor)
```

```python
gray = cv2.cvtColor(child_img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray)
```

Writing the image

```python
cv2.imwrite('grayscale_child.png', gray)
```
