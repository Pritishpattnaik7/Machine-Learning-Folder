```python
#importing openCV
import cv2
import matplotlib.pyplot as plt
```

```python
#reading images
img1 = cv2.imread("dog.png")

#for original colour
new_img = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img1)
plt.show()
plt.imshow(new_img)
plt.show()
```
