```python
#matplotlib -> used for data visualization
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1,2,3]
y = [2,5,8]
plt.scatter(x, y)
plt.show()
```

```python
x = [1,2,3]
y = [2,5,8]
plt.scatter(x,y)
plt.plot(x, y)
plt.show()
```

```python
#plotting 2 lines

x = [1,2,3]
y = [2,5,8]
x1 = [4,6,9]
y1 = [5,6,7]

plt.scatter(x,y)
plt.plot(x,y)

plt.scatter(x1,y1)
plt.plot(x1, y1)
plt.show()
```

```python
#changing colour of lines
#plot has the 3rd argument to change the colour and symbol of line("colour1stletter character")

x = [1,2,3]
y = [2,5,8]
x1 = [4,6,9]
y1 = [5,6,7]

plt.plot(x,y)
plt.plot(x1, y1, "r+")
```

```python
#plot function x^3
x = np.array([1,2,3,4])
y = x ** 3
plt.plot(x, y)
plt.show()
```

```python
a = np.arange(0, 5, 0.1)
b = a ** 3
plt.plot(a, b, "b.")
plt.show()
```

```python
c = [3,4,5]
plt.plot(c)
plt.show()
```

```python
#Customizing the graphs
#color , marker , linewidth, label , ylabel & xlabel , legend() , grid(), text()
 
a = np.arange(0, 5, 0.1)
b = a ** 3
c = a ** 2


plt.plot(a, b, color = 'orange', linewidth = 4, label = "x^3") #we can also add marker
plt.plot(a, c, color = 'black' , linewidth = 3, label = "x^2")

plt.ylabel("y")
plt.xlabel("x")

plt.legend()   #prints the graph info(top LHS)
plt.grid()     #creates the grid

plt.text(2,80, "Text")
plt.text(4, 60, '(4,60)')  #here we can add text a a given co-ordinate

plt.title("Matplotlib demo")
plt.show()
```

```python
#Pie chart

size = np.array([35, 25, 25, 15])
myLabels = ["Apples", "Bananas", "Grapes", "Oranges"]
myColours = ["red", "yellow", "purple", "orange"]
plt.pie(size, labels=myLabels , colors=myColours)
plt.show()
```
