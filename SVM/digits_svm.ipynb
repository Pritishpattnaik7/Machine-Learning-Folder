```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

```python
dir(digits)
```

```python
digits.data
```

```python
digits.target_names
```

```python
digits.target
```

```python
df = pd.DataFrame(digits.data, columns=digits.feature_names)
```

```python
df.head()
```

```python
df['target'] = digits.target
df.head()
```

```python
#seperating the dataframe into x and y
```

```python
x = df.drop(['target'], axis='columns')
x
```

```python
y = df.target
y
```

```python
#model selection train and test
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
#training the model
```

```python
from sklearn.svm import SVC
```

```python
#using the default rbf kernel
rbf_clf = SVC(kernel='rbf')
```

```python
rbf_clf.fit(x_train, y_train)
```

```python
y_pred_test = rbf_clf.predict(x_test)
print(y_pred_test)
```

```python
rbf_clf.score(x_train, y_train)
```

```python
rbf_clf.score(x_test, y_test)
```
