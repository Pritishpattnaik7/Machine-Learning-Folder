```python
from sklearn import datasets
```

```python
boston = datasets.load_boston()
boston
```

```python
x = boston.data
y = boston.target
```

```python
type(x)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(x)
```

```python
df
```

```python
boston.feature_names
```

```python
df.columns = boston.feature_names
```

```python
df
df["age_age"] = df.AGE**2
x2 = df.values
x2
df.describe()

df["crim^2"] = df.CRIM**2
df["zn^2"] = df.ZN**2
df["indus^2"] = df.INDUS**2
df["chas^2"] = df.CHAS**2
df["nox^2"] = df.NOX**2
df["rm^2"] = df.RM**2
df["dis^2"] = df.DIS**2
df["rad^2"] = df.RAD**2
df["tax^2"] = df.TAX**2
df["ptratio^2"] = df.PTRATIO**2
df["b^2"] = df.B**2
df["lstat^2"] = df.LSTAT**2

x3 = df.values
df.describe()
```

```python
from sklearn import model_selection
```

```python
x_train, x_test, y_train, y_test = model_selection.train_test_split(x, y, random_state=0)
x2_train, x2_test, y2_train, y2_test = model_selection.train_test_split(x2, y, random_state=0)
x3_train, x3_test, y3_train, y3_test = model_selection.train_test_split(x3, y, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
algo1 = LinearRegression()
algo2 = LinearRegression()
algo3 = LinearRegression()
```

```python
algo1.fit(x_train, y_train)
algo2.fit(x2_train, y2_train)
algo3.fit(x3_train, y3_train)
```

```python
y_pred = algo1.predict(x_test)
```

```python
train_score = algo1.score(x_train, y_train)
test_score = algo1.score(x_test, y_test)

train2_score = algo2.score(x2_train, y2_train)
test2_score = algo2.score(x2_test, y2_test)

train3_score = algo3.score(x3_train, y3_train)
test3_score = algo3.score(x3_test, y3_test)
```

```python
print("Training score : ",train_score)
print("Testing score : ", test_score)

print("Training2 score : ",train2_score)
print("Testing2 score : ", test2_score)

print("Training3 score : ",train3_score)
print("Testing3 score : ", test3_score)
```
